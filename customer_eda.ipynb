```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Data cleaning

```python
df.describe()
```

```python
#  no duplicates
df.drop_duplicates().count()
```

## Handle missing values

```python
# ~17% of signup_date, ~7% of satisfaction_score, 4% payment_method are missing
# ~1.1% of regions may be recoverable from city_raw
df.isna().sum() *100/df.count()
```

```python
# signup_date missingness is evenly distributed across churned/non-churned
# and shows no pattern with spend or sessions — likely missing at random
df.groupby(df["signup_date"].isna())["churned"].mean()
df.groupby(df["signup_date"].isna())["monthly_sessions"].mean()
df.groupby(df["signup_date"].isna())["avg_monthly_spend"].mean()
df[df["signup_date"].isna()]["acquisition_channel"].value_counts()
```

```python
# Non-churned customers are slightly more likely to have missing satisfaction scores (6.5% vs 3.4%)
# Missingness is not completely random — a flag column is added before imputation
df.groupby(df["satisfaction_score"].isna())["churned"].mean()
df["satisfaction_missing"] = df["satisfaction_score"].isna().astype(int)
df["satisfaction_score"] = df["satisfaction_score"].fillna(
    df.groupby("churned")["satisfaction_score"].transform("median")
)
df["satisfaction_score"].unique()
```

```python
# Missing signup_date doesnt depend on churned status
df.groupby('churned')['signup_date'].apply(lambda x: x.isna().mean())*100
```

```python
df["acquisition_channel"].unique()
```

```python
df["payment_method"] = df["payment_method"].fillna("unknown")
df["acquisition_channel"] = df["acquisition_channel"].fillna("unknown")
```

```python
df=df.dropna(subset="estimated_income_czk")
df=df.dropna(subset="product_plan")
```

```python
# Likely means no tickets
df["support_tickets_90d"] = df["support_tickets_90d"].fillna(0)
```

## Invalid values

```python
# Removed 12 customers with negative age and 13 over 110 — likely data entry errors
df[df["age"]<0].count() 
df[df["age"]>110].count()
df = df[df["age"]>0]
df = df[df["age"]<110]
```

```python
# Removed 7 customers with last_active_date before 1950 — invalid timestamps
df["last_active_date"] = pd.to_datetime(df["last_active_date"], errors="coerce")
df[df["last_active_date"].dt.year <1950].count()

df = df[df["last_active_date"].dt.year >1950]
```

```python
# Removed 17 customers with negative avg_monthly_spend — invalid values
df[df["avg_monthly_spend"]<0].count()
df = df[df["avg_monthly_spend"]>0]
```

## Correct data types

```python
df.dtypes
```

```python
for col_name in df.columns:
    if (df[col_name].dtype == "str"):
        df[col_name] = df[col_name].str.strip().str.lower()
```

```python
df["age"] = pd.to_numeric(df["age"]).astype("int64")
df["monthly_sessions"] = pd.to_numeric(df["monthly_sessions"]).astype("int64")
df["estimated_income_czk"] = pd.to_numeric(df["estimated_income_czk"], errors="coerce").astype("Int64")
```

```python
df["payment_method"].drop_duplicates().sort_values(ascending=False)
```

```python
df["support_tickets_90d"]=df["support_tickets_90d"].astype("Int32")
```

## Handle missing regions

```python
df["city_raw"].value_counts()
```

```python
df[df["city_raw"] == "unknown"]
```

```python
city_to_region = (
    df.dropna(subset=["region"])
      .drop_duplicates(subset=["city_raw"])
      .set_index("city_raw")["region"]
)

df["region"] = df["region"].fillna(df["city_raw"].map(city_to_region))
city_to_region
```

```python
city_mapping = {
    "praha": "prague",
    "prg": "prague",
    "prague": "prague",

    "brno": "brno",
    "brn": "brno",

    "olomouc": "olomouc",

    "ostrava": "ostrava",

    "plzen": "plzen",
    "plzeň": "plzen",
    "pilsen": "plzen",

    "liberec": "liberec",

    "unknown": "unknown",
    
    "other": "other"
}

df["city_clean"] = df["city_raw"].replace(city_mapping)
```

# Correlations

```python
df_numerized = df.copy()
for col_name in df_numerized:
    if (df_numerized[col_name].dtype == 'string'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='spearman').round(1)
sns.heatmap(correlation_matrix, annot =True)
plt.title('Correlation matrix for all features')
plt.xlabel('customers features')
plt.ylabel('customers features')
plt.show()
```

```python
# last_active_date has the strongest correlation with churn
# support_tickets_90d, satisfaction_score, product_plan and monthly_sessions are also worth investigating
corr =df_numerized.corr(method='spearman')
churn_corr = corr["churned"]
corr_pairs = churn_corr.sort_values()
high_corr = corr_pairs[(corr_pairs > 0.1) | (corr_pairs < -0.1)]
high_corr
```

```python
corr_pairs = corr.unstack()
corr_pairs = corr_pairs.sort_values()
corr_pairs
```

```python
# Notable feature pairs:
# support_tickets_90d - satisfaction_score (-0.50): more issues = lower satisfaction
# monthly_sessions - avg_monthly_spend (0.82): more monthly sessioins = higher spend
# product_plan - avg_monthly_spend (0.56): higher plans spend more
# monthly_sessions - product_plan (0.44): higher plans have more monthly sessions

high_corr = corr_pairs[(corr_pairs > 0.4) | (corr_pairs < -0.4)]
high_corr = high_corr[high_corr < 1]
high_corr
```

# Exploration

## Churned

### Last active date and churned status

```python
# Most churned customers had their last activity in January 2026
year = df.groupby(df["last_active_date"].dt.year)["churned"].sum()
year.plot(kind="bar")
plt.xlabel("Year")
plt.ylabel("Churned customers (last activity count)")
plt.title("Churned customers by last activity year")

plt.show()
```

```python
# suggesting a possible seasonal pattern or a product/service change at that time
df_2026 = df[df["last_active_date"].dt.year == 2026]
month = df_2026.groupby(df_2026["last_active_date"].dt.month)["churned"].sum()
month.plot(kind="bar")
plt.xlabel("month")
plt.ylabel("count of churned customers")
plt.title("Churned customers in 2026 by months")
```

### Support tickets and churned status

```python
# Customers with more support tickets are significantly more likely to churn
# median ticket count is higher for churned customers across all plan types
sns.boxplot(x="churned", y="support_tickets_90d", data =df)
plt.xlabel("churned")
plt.ylabel("count of support tickets")
plt.title("boxplot of support tickets count to churned customers")
plt.show()
```

### Satisfaction score and churned status

```python
# Churned customers report slightly lower satisfaction scores
# the difference is modest but consistent — satisfaction alone is not a strong standalone predictor
sns.boxplot(x="churned", y="satisfaction_score", data=df)
plt.ylim(0, 20) 
plt.xlabel("Churned")
plt.ylabel("Avg satisfaction score")
plt.title("Average satisfaction by churn")
plt.show()
```

### Product plan and churned status

```python
# Basic plan customers churn at ~5.7% — 3-5x higher than plus, premium and business plans
# Higher-tier customers show strong retention regardless of spend level
df.groupby('product_plan')['churned'].mean()*100
```

### Average monthly sessions and churned status

```python
# Churn rate drops steadily as session frequency increases
# customers with 0-8 sessions churn at 5.1%; above 24 sessions, churn reaches 0%
# session frequency is the strongest behavioral predictor of retention in this dataset
(df.groupby('monthly_sessions')['churned'].mean()*100).plot(kind="line")
plt.ylabel( "% of churned clients")
plt.xlabel("Monthly sessions")
plt.title("% of churned to avg month session count")
plt.show()
```
